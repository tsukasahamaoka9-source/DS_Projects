# アンサンブル
Random Forestと同じように、いろんな手法の結果をまとめる

今後のテストデータの更新など、対応に幅を持たせる意味で重要。

```python
# ファイルの読み込みと名前付け
submit1 = pd.read_csv('/kaggle/input/datasets/tsukasahamaoka/ensemble-resources/Kfold_submit.csv')
submit2 = pd.read_csv('/kaggle/input/datasets/tsukasahamaoka/ensemble-resources/randomforest_reenginerring_submit.csv')

# 各値を確認
submit_all = pd.DataFrame({'submit1': submit1['Survived'].values,
                            'submit2': submit2['Survived'].values,})

submit_all.head()
```

```python
submit_all.corr()
# なるべく似たような相関を持つつまり似たようなアルゴリズムはアンサンブルしない
```

```python
# 加算
submit_ensemble = pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
submit_ensemble['Survived'] = submit1['Survived'] + submit2['Survived']

submit_ensemble.tail()
```

```python
# 結果が2以上なら生存とみなす
submit_ensemble['Survived'] = (submit_ensemble['Survived'] >= 2).astype(int)
submit_ensemble.to_csv("ensemble_submit.csv", index = False)
```
